# Subtask 3.6 Find popular genres

```python
# genres columns in the dataframe with all the genres of the movies sepearted by a pipe(|).out of all the movies genres,the first two 
# two are most significant for any film.
# 1. Extract the first two genres from the genres column and store them in two new columns: genre_1 and genre_2.some of the movies
#    might have only one genre.In such situtaion,extract the single genre nto both the columns,for such movies the genre_2 will be 
#    the same as genre_1.
# 2. Group the dataframe using genre_1 as the primray column and genre_2 as the secondary column.
# 3. Find out the most popular combo of genres by finding the mean of the gross values using the gross column and store them in
#    a new dataframe name Popgenre.
```

```python
tempgenre=movies.genres.str.split("|",expand=True).iloc[:,:2]
tempgenre.columns=["genre_1","genre_2"]
```

```python
tempgenre
```

```python
tempgenre.genre_2.fillna(tempgenre.genre_1,inplace=True)
tempgenre
```

```python
movies=pd.concat([movies,tempgenre],axis=1)
movies
```

```python
movies.groupby(["genre_1","genre_2"]).gross.mean().sort_values(ascending=False).head(5)
```
